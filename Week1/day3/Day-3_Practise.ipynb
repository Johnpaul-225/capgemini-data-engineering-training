### Salary Hike Based on Experience and Performance

```sql
select *,
case
  when experience >=8 and performance_rating =='A' then '20% hike'
  when experience >=5 and performance_rating =='B' then '15% hike'
  when performance_rating =='C' then 'NO hike'
  else '10% hike'
end as Hike
from Employee
```

```sql
select *,
case
  when experience >=8 and performance_rating =='A' then salary*1.20
  when experience >=5 and performance_rating =='B' then salary*1.15
  when performance_rating =='C' then salary
  else salary*1.10
end as Hiked_Salary
from Employee
```

## Problem 2: Bonus Calculation Based on Department and Performance

```sql
select *,
case
  when department ='Finance' and performance_rating='A' then '20% bonus'
  when department ='Finance' and performance_rating='B' then '15% bonus'
  when department ='Finance' and performance_rating='C' then '5% bonus'
  when department ='Engineering' and performance_rating="A" then '18% bounus'
  when department ='Engineering' and performance_rating="B" then '12% bonus'
  when department ='Engineering' and performance_rating="C" then '3% bonus'
  else '10% bonus'
end as Bonus
from Employee
```

```sql
select *,
case
  when department ='Finance' and performance_rating='A' then 0.20*salary
  when department ='Finance' and performance_rating='B' then 0.15*salary
  when department ='Finance' and performance_rating='C' then 0.05*salary
  when department ='Engineering' and performance_rating="A" then 0.18*salary
  when department ='Engineering' and performance_rating="B" then 0.12*salary
  when department ='Engineering' and performance_rating="C" then 0.03*salary
  else 0.10*salary
end as Bonus_Salary
from Employee
```

## Categorizing Employees by Salary Range and Performance

```sql
select *,
case
  when salary > 80000 and performance_rating='A' then 'High Performer'
  when salary between 50000 and 80000 and performance_rating='B' then 'Mid Performer'
  when salary < 50000 or performance_rating='C' then 'Low Performer'
  else 'Uncategorized'
end as Performer
from Employee
```

## Risk Assessment Based on Experience and Department

```sql
select *,
case
  when department='HR' and experience < 5 then 'High Risk'
  when department='HR' and experience >5 then 'Low Risk'
  when department in ('Engineering', 'Finance') and experience > 8 then 'Low Risk'
  when department in ('Engineering', 'Finance') and experience < 8 then 'High Risk'
  else 'Medium Risk'
end as Risk
from Employee
```

#####Problem 1: Nested CASE for Performance and Salary Hike Based on Multiple Criteria
######Scenario: You want to determine the salary hike based on performance rating, experience, and current salary. 


```sql
select *,
case 
  when performance_rating='A' then
    case
      when salary > 80000 and experience > 5 then '25% hike'
      when salary > 80000 then '20% hike'
      when salary between 50000 and 80000 then '15% hike'
      else 'salary'
    end
  when performance_rating='B' then
    case
      when experience > 5 then '12% hike'
      else '10% hike'
    end
  when performance_rating='C' then 'No hike'
  else 'no hike'
end as Hike
from Employee
```

```sql
select *,
case 
  when performance_rating='A' then
    case
      when salary > 80000 and experience > 5 then salary*1.25
      when salary > 80000 then salary*1.20
      when salary between 50000 and 80000 then salary*1.15
      else salary
    end
  when performance_rating='B' then
    case
      when experience > 5 then salary*1.12
      else salary*1.10
    end
  when performance_rating='C' then salary
  else salary
end as Hike
from Employee
```

### 2.Department and Performance
#####Scenario: You are tasked with giving bonuses based on the department, performance rating, and experience:


```sql
select *,
case 
  when department='Finance' then
    case 
      when performance_rating='A' and experience > 10 then '25% Bonus'
      else '20% Bonus'
    end
  when department='HR' then
    case 
      when performance_rating='B' or experience > 5 then '15% Bonus'
      else '10% Bonus'
    end
  else "8% Bonus" 
end as Bonus
from Employee
```

```sql
SELECT *,
CASE 
  WHEN department = 'Finance' AND performance_rating = 'A' AND experience > 10 THEN salary * 0.25
  WHEN department = 'Finance' THEN salary * 0.20
  WHEN department = 'HR' AND (performance_rating = 'B' OR experience > 5) THEN salary * 0.15
  WHEN department = 'HR' THEN salary * 0.10
  ELSE salary * 0.08
END AS bonus_amount
FROM Employee;
```

## Problem 4: Risk Assessment Based on Experience and Department

```sql
select *,
case 
  when department='HR' then
    case
      when experience > 5 then 'Low Risk'
      else 'High Risk'
    end
  when department in ('Engineering' , 'Finance') then 
    case
      when experience > 8 then 'Low Risk'
      else 'Medium Risk'
    end
  else 'Medium Risk'
end as risk_level
from Employee
```
