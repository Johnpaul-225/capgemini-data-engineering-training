### 17. List the number of sales made by each employee.

### 19. Calculate the average sales amount grouped by product.

```sql
select product, avg(amount) as avg_sales
from sales
group by product
```

### 20. Find employees who have made more than 2 sales, grouped by their names.

```sql
select E.emp_name, count(S.sales_id) as no_of_sales
from Employee E
inner join sales S
on E.emp_id = S.emp_id
group by E.emp_name
having count(S.sales_id) > 2
```

### 21. Calculate the total salary and total sales amount for each employee.

```sql
select E.emp_name, E.salary as tot_salary, sum(S.amount) as tot_amount
from Employee E
inner join sales S
on E.emp_id = S.emp_id
group by E.emp_name,E.salary
```

### 22. Count the number of unique products sold by each employee.

```sql
select E.emp_name, count(Distinct(S.product)) as no_of_products
from Employee E
left join sales S
on E.emp_id = S.emp_id
group by E.emp_name
```

### 23. Find the highest sales amount made by each employee.

```sql
select E.emp_name, max(S.amount) as highest_sales
from Employee E
inner join sales S
on E.emp_id = S.emp_id
group by E.emp_name
```

### 24. Calculate the total sales amount grouped by product and filtered by products where the total exceeds 50,000.


```sql
SELECT 
    product,
    SUM(amount) AS total_sales
FROM Sales
GROUP BY product
HAVING SUM(amount) > 50000;
```

### 25. Find the departments with the highest average sales amount. Real-World Scenarios


```sql
select E.department, avg(S.amount) as avg_sales
from Employee E
inner join sales S
on E.emp_id = S.emp_id
group by E.department
```

### 26. Find the department with the highest total sales amount.

```sql
select E.department, sum(S.amount) as tot_sales
from Employee E
inner join sales S
on E.emp_id = S.emp_id
group by E.department
order by tot_sales desc
limit 1
```

### 27. Show the top 3 employees with the highest total sales amount, grouped by employee names.


```sql
select E.emp_name, E.emp_id, sum(S.amount) as highest_sales
from Employee E
inner join sales S
on E.emp_id = S.emp_id
group by E.emp_name, E.emp_id
order by highest_sales desc
limit 3
```

30. Show employees who have not made any sales, grouped by their department

```sql
SELECT E.department,COUNT(E.emp_id) AS no_sales_employees
FROM Employee E
LEFT JOIN Sales S
ON E.emp_id = S.emp_id
WHERE S.emp_id IS NULL
GROUP BY E.department;
```
