## Displaying Table

```sql
select * from regex_practice
```

#### From the column mixed_value, extract the numeric characters that appear at the very beginning of the value.
#### The extraction must stop as soon as a non-numeric character appears.


```sql
select mixed_value, regexp_extract(mixed_value, '^[0-9]+',0) as number
from regex_practice
```

```sql
select mixed_value, nullif(regexp_extract(mixed_value, '^[0-9]+',0),'') as number
from regex_practice
```

#### From the column mixed_value, extract the numeric characters that appear at the very end of the value.
#### No alphabetic or special characters should be included.


```sql
select mixed_value, regexp_extract(mixed_value, '[0-9]$',0) as ending_number
from regex_practice
```

### From the column mixed_value, extract only the first single character of the value, regardless of whether it is a letter or a number.


```sql
select mixed_value, regexp_extract(mixed_value, '^.',0) as first_single
from regex_practice
```

### From the column mixed_value, extract only the last single character of the value.

```sql
select mixed_value, regexp_extract(mixed_value, '.$',0) as last_single
from regex_practice
```

#### From the column mixed_value, extract exactly two consecutive numeric characters that appear anywhere in the value.
#### Do not extract more or fewer than two digits.


```sql
select mixed_value, regexp_extract(mixed_value, '[0-9]{2}',0) as two_digits
from regex_practice
```

### From the column mixed_value, extract exactly one numeric character that appears anywhere in the value.


```sql
select mixed_value, regexp_extract(mixed_value, '[0-9]',0) as one_digit
from regex_practice
```

#### From the column phone, extract the country code present at the beginning of the phone number.  The extracted value must contain only the country code digits.

```sql
select phone, regexp_extract(phone, '^\\+?([0-9]+)-',1) as Country_code
from regex_practice
```

### From the column mixed_value, extract the numeric portion that is present between alphabetic characters.


```sql
select mixed_value, regexp_extract(mixed_value, '[a-zA-Z]([0-9]+)[a-zA-Z]',1) as number
from regex_practice
```

### From the column email, extract the text that appears before the at-symbol. The extracted value must not include the at-symbol itself.


```sql
select email, regexp_extract(email, '^([^@]+)',1) as username
from regex_practice
```

### From the column email, extract the text that appears after the at-symbol including the domain name.


```sql
select email, regexp_extract(email, '@(.+)',1) as domain
from regex_practice
```

### From the column email, extract only the domain name without including the at-symbol.

```sql
select email, regexp_extract(email, '@([^@]+)',1) as domain
from regex_practice
```

### From the column email, extract only the text that appears after the last dot in the email address.

```sql
select email, regexp_extract(email, '\\.([^.]+)',1) as extension
from regex_practice
```

### From the column mixed_value, extract only alphabetic characters that appear together as a continuous sequence.


```sql
select mixed_value, regexp_extract(mixed_value, '[a-zA-z]+',0) as character
from regex_practice
```

### From the column full_text, extract exactly the first three characters of the value.

```sql
select full_text, regexp_extract(full_text, '^.{3}',0) as first_three_char
from regex_practice
```

### From the column full_text, extract exactly the last two characters of the value.

```sql
select full_text, regexp_extract(full_text, '.{2}$',0) as last_two_char
from regex_practice
```

### From the column full_text, extract the employee number portion that appears between the alphabetic prefix and the first underscore.


```sql
SELECT full_text, REGEXP_EXTRACT(full_text, '^[A-Za-z]+([0-9]+)_', 1) AS emp_number
FROM regex_practice;
```

### From the column full_text, extract the country code that appears at the end of the value.

```sql
select full_text, regexp_extract(full_text, '_([^_.]+)',1) as emp_number
from regex_practice
```

```sql
SELECT full_text, REGEXP_EXTRACT(full_text, '_([0-9]+)$', 1) AS country_code
FROM regex_practice;
```

### From the column full_text, extract the alphabetic text that appears between two underscore characters.


```sql
SELECT full_text, REGEXP_EXTRACT(full_text, '_([A-Za-z]+)_', 1) AS country_text
FROM regex_practice;
```

#### From the column phone, extract the numeric characters that appear immediately after the plus sign. The extraction must include only the digits that represent the country code.


```sql
SELECT phone, REGEXP_EXTRACT(phone, '\\+([0-9]+)', 1) AS country_code
FROM regex_practice;
```
