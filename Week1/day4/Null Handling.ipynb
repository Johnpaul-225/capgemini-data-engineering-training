### Replace NULL salary with 0

```sql
select coalesce(salary,0) as salary
from employees
```

### Replace NULL bonus with 1000

```sql
select coalesce(bonus,1000) as bonus
from employees
```

### Show order amount, if NULL replace with 500

```sql
select coalesce(amount,500)
from orders
```

### Replace NULL stock with 0

```sql
select coalesce(stock,0) as stock
from products
```

### Show employee earnings using: salary, if NULL use bonus


```sql
select emp_id, name, bonus, coalesce(salary,bonus) as salary
from employees
```

### Show first available value: salary → bonus → 0


```sql
select coalesce(salary,bonus,0) as amount
from employees
```

### Show product price: price → 1000 (default)


```sql
select coalesce(price,1000) as price
from products
```

### Get customer payment: amount → discount → 0

```sql
select amount,discount,coalesce(amount,discount,0) as total
from orders
```

### Convert salary to NULL if salary = 0

```sql
select salary, nullif(salary,0) as updated_salary
from employees
```

### Convert discount to NULL if discount = 0

```sql
select discount, nullif(discount,0) as updated_discount
from orders
```

### Use NULLIF to avoid divide by zero:

```sql
SELECT amount / NULLIF(discount, 0)
FROM Orders;
```

### Replace coupon_code with NULL if it is 'DISC10'

```sql
SELECT NULLIF(coupon_code, 'DISC10') AS coupon_code
FROM Orders;
```

### Calculate total earnings: salary + bonus (handle NULL properly)


```sql
SELECT COALESCE(salary, 0) AS salary, COALESCE(bonus, 0) AS bonus, COALESCE(salary, 0) + COALESCE(bonus, 0) AS total
FROM employees;
```

### Show employee income: COALESCE(salary, bonus, 1000)


```sql
select name, salary, bonus, coalesce(salary,1000) as updated_salary, coalesce(bonus,1000) as updated_bonus, 
coalesce(salary,1000)+coalesce(bonus,1000) as income
from employees
```

### Replace empty discount with NULL using NULLIF: NULLIF(discount, 0)


```sql
select nullif(discount,0) as discount
from orders 
```

### Show final payable amount: amount - discount (handle NULL)


```sql
select coalesce(amount,0) as amount, coalesce(discount,0) as discount,
coalesce(amount,0)-coalesce(discount,0) as final_amount
from orders
```

```sql
SELECT COALESCE(amount, 0) AS amount, COALESCE(discount, 0) AS discount,
GREATEST(COALESCE(amount, 0) - COALESCE(discount, 0), 0) AS final_amount
FROM orders;
```
