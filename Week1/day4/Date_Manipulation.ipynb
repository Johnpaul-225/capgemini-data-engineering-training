## Creating Order table and Inserting values 

```sql
DROP TABLE IF EXISTS orders;
CREATE TABLE orders (
order_id INT PRIMARY KEY,
customer_name VARCHAR(50),
order_date DATE,
order_timestamp TIMESTAMP,
delivery_date DATE,
order_amount DECIMAL(10,2)
);

INSERT INTO orders VALUES
(1, 'Karthik', '2024-01-15', '2024-01-15 10:30:45', '2024-01-20', 2500.00),
(2, 'Veena', '2024-02-18', '2024-02-18 18:45:20', '2024-02-22', 3200.50),
(3, 'Ravi', '2024-03-02', '2024-03-02 09:15:10', '2024-03-08', 4100.75),
(4, 'Anil', '2024-03-09', '2024-03-09 14:05:55', '2024-03-15', 1800.00),
(5, 'Suresh', '2024-01-07', '2024-01-07 23:55:00', '2024-01-12', 2900.00);
```

### EXTRACTING YEAR, MONTH, DAY

```sql
SELECT YEAR(order_date)as Year, MONTH(order_date) as Month, DAY(order_date) as Day
FROM orders;
```

### EXTRACT USING EXTRACT()

```sql
SELECT EXTRACT(YEAR FROM order_date) as Year, EXTRACT(MONTH FROM order_date) as Month,
EXTRACT(DAY FROM order_date) as Day
FROM orders;
```

### MONTH NAME AND DAY NAME

```sql
SELECT MONTHNAME(order_date) as Month, DAYNAME(order_date) as Day
FROM orders
```

### WEEKDAY AND DAY OF WEEK

```sql
SELECT WEEKDAY(order_date) as weekDay, DAYOFWEEK(order_date) as dayOfWeek
FROM orders
```

### IDENTIFY WEEKENDS (SATURDAY & SUNDAY)

```sql
select order_id, order_date
from orders
where dayname(order_date) in ('saturday', 'sunday')
```

```sql
SELECT order_id, order_date
FROM orders
WHERE DAYOFWEEK(order_date) IN (1, 7);
```

```sql
SELECT order_id, order_date
FROM orders
WHERE weekday(order_date) BETWEEN 5 AND 6
```

### IDENTIFY WEEKDAYS

```sql
SELECT order_id, order_date
FROM orders
WHERE WEEKDAY(order_date) BETWEEN 2 AND 6
```

```sql
SELECT order_id, order_date
FROM orders
WHERE weekday(order_date) BETWEEN 0 AND 4
```

### DATE ARITHMETIC

### Add / Subtract Days

```sql
SELECT order_date, DATE_ADD(order_date, 5) as new_date
FROM orders;
```

```sql
select order_date, date_sub(order_date,5) as new_date
from orders
```

### Add / Subtract Months

```sql
select order_date, add_months(order_date, 5) as new_date
from orders
```

```sql
select order_date, add_months(order_date, -5) as new_date
from orders
```

### Add / Subtract Years

```sql
select order_date, add_months(order_date, 12*2) as new_date
from orders
```

```sql
select order_date, add_months(order_date, -12*2) as new_date
from orders
```

### TIMESTAMPDIFF (Difference in Units)

```sql
SELECT
TIMESTAMPDIFF(DAY, order_date, delivery_date) AS days_diff,
TIMESTAMPDIFF(MONTH, order_date, delivery_date) AS months_diff
FROM orders;
```

### LAST DAY OF MONTH

```sql
SELECT LAST_DAY(order_date) FROM orders
```

### FIRST DAY OF MONTH

```sql
SELECT order_date, date_sub(order_date, day(order_date) - 1) AS first_day
FROM orders;
```

## DATE FORMAT 

```sql
SELECT order_date, date_format(order_date, 'dd-MM-yyyy') 
FROM orders;
```

```sql
SELECT order_date, date_format(order_date, 'MM-dd-yyyy') 
FROM orders;
```

### STRING TO DATE 

```sql
SELECT TO_DATE('15-01-2024', 'd-mm-yyyy') as date
```

### TIMESTAMP FORMATTING

```sql
SELECT date_format(order_timestamp, 'dd-MM-yyyy hh:mm:ss') as new_date_format
FROM orders;
```

### FILTER BY MONTH (JAN, FEB, MAR)

```sql
SELECT *
FROM orders
WHERE month(order_date) IN (1, 2, 3);
```

### FINANCIAL YEAR LOGIC 

```sql
SELECT order_date,
CASE
  WHEN MONTH(order_date) >= 4 THEN CONCAT(YEAR(order_date), '-', YEAR(order_date)+1)
  ELSE CONCAT(YEAR(order_date)-1, '-', YEAR(order_date))
END AS financial_year
FROM orders;
```

### Orders in Last 7 Days

```sql
SELECT *
FROM orders
WHERE order_date >= CURDATE() - 7
```

### Orders Placed Today

```sql
SELECT *
FROM orders
WHERE DATE(order_timestamp) = CURDATE();
```
