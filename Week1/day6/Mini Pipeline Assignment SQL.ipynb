```sql
SELECT s.*
FROM sales s
LEFT ANTI JOIN customers c
ON s.customer_id = c.customer_id
```

```sql
SELECT sd.*
FROM sales_dealer sd
LEFT ANTI JOIN dealers d
ON sd.dealer_id = d.dealer_id
```

## 2. Validation Report

```sql
SELECT 'Missing Customer' AS issue, COUNT(*) AS count
FROM sales s
LEFT JOIN customers c ON s.customer_id = c.customer_id
WHERE c.customer_id IS NULL

UNION ALL

SELECT 'Missing Car', COUNT(*)
FROM sales s
LEFT JOIN cars c ON s.car_id = c.car_id
WHERE c.car_id IS NULL

UNION ALL

SELECT 'Missing Dealer', COUNT(*)
FROM sales_dealer sd
LEFT JOIN dealers d ON sd.dealer_id = d.dealer_id
WHERE d.dealer_id IS NULL

UNION ALL

SELECT 'Negative Price', COUNT(*)
FROM cars
WHERE price < 0

UNION ALL

SELECT 'Null Customer Name', COUNT(*)
FROM customers
WHERE name IS NULL

UNION ALL

SELECT 'Invalid Quantity', COUNT(*)
FROM sales
WHERE quantity <= 0;
```

# Transformation

## 1. Revenue per Customer

```sql
SELECT c.customer_id, c.name, SUM(car.price * s.quantity) AS total_revenue
FROM sales s
JOIN customers c 
ON s.customer_id = c.customer_id
JOIN cars car 
ON s.car_id = car.car_id
GROUP BY c.customer_id, c.name
ORDER BY total_revenue DESC
```

## 2 Cars per Brand

```sql
SELECT brand, COUNT(*) AS total_cars
FROM cars
GROUP BY brand
ORDER BY total_cars DESC;
```

## 3. City Revenue

```sql
SELECT c.city, SUM(car.price * s.quantity) AS city_revenue
FROM sales s
JOIN customers c 
ON s.customer_id = c.customer_id
JOIN cars car 
ON s.car_id = car.car_id
GROUP BY c.city
ORDER BY city_revenue DESC;
```

# Dealer Analytics

## 1. Revenue per dealer

```sql
CREATE OR REPLACE TEMP VIEW dealer_base AS
SELECT d.dealer_id, d.name AS dealer_name, d.city AS dealer_city, car.price, s.quantity
FROM sales s
JOIN cars car ON s.car_id = car.car_id
JOIN sales_dealer sd ON s.sale_id = sd.sale_id
JOIN dealers d ON sd.dealer_id = d.dealer_id;
```

## 1. Revenue per dealer

```sql
SELECT dealer_id, dealer_name, dealer_city, SUM(price * quantity) AS total_revenue
FROM dealer_base
GROUP BY dealer_id, dealer_name, dealer_city
```

## 2) Top Dealers by Revenue

```sql
SELECT dealer_id, dealer_name, dealer_city, SUM(price * quantity) AS total_revenue
FROM dealer_base
GROUP BY dealer_id, dealer_name, dealer_city
order BY total_revenue DESC
```

## 3. Dealer-City Contribution

```sql
SELECT d.dealer_id, d.name AS dealer_name, d.city AS dealer_city,
    SUM(car.price * s.quantity) AS total_revenue,
    ROUND(
        SUM(car.price * s.quantity) * 100.0 /
        SUM(SUM(car.price * s.quantity)) OVER (PARTITION BY d.city),
        2
    ) AS contribution_pct
FROM sales s
JOIN cars car 
    ON s.car_id = car.car_id
JOIN sales_dealer sd 
    ON s.sale_id = sd.sale_id
JOIN dealers d 
    ON sd.dealer_id = d.dealer_id
GROUP BY d.dealer_id, d.name, d.city;
```

# SQL Analysis

## 1. Top 3 Customers per City

```sql
SELECT *
FROM (
    SELECT c.city, c.customer_id, c.name,
        SUM(car.price * s.quantity) AS revenue,
        RANK() OVER (
            PARTITION BY c.city 
            ORDER BY SUM(car.price * s.quantity) DESC
        ) AS rnk
    FROM sales s
    JOIN customers c ON s.customer_id = c.customer_id
    JOIN cars car ON s.car_id = car.car_id
    GROUP BY c.city, c.customer_id, c.name
) 
WHERE rnk <= 3;
```

## 2. Repeat Customers

```sql
SELECT customer_id, COUNT(*) AS total_orders
FROM sales
GROUP BY customer_id
HAVING COUNT(*) > 1;
```

## 3. Monthly Revenue Trend

```sql
SELECT 
DATE_FORMAT(s.sale_date, 'MM') AS month,
SUM(car.price * s.quantity) AS revenue
FROM sales s
JOIN cars car ON s.car_id = car.car_id
GROUP BY DATE_FORMAT(s.sale_date, 'MM')
ORDER BY month;
```
