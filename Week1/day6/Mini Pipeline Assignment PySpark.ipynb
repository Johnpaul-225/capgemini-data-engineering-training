# Cleaning

## 1. Handle Nulls + Trim

```python
customers = customers.withColumn("name", trim(col("name"))) \
                     .fillna({"name": "Unknown"})
display(customers)
```

## 2. Fix Negative Price

```python
cars = cars.withColumn("price",when(col("price") < 0, None).otherwise(col("price")))
display(cars)
```

## 3. Remove Invalid Keys

```python
sales_clean = sales.join(customers, "customer_id", "inner") \
                   .join(cars, "car_id", "inner")
display(sales_clean)
```

## Validation

## 1. Anti Join

```python
invalid_customers = sales.join(customers, "customer_id", "left_anti")
invalid_customers.show()
```

## 2. Validation Report

```python
validation = spark.createDataFrame([
    ("Null Names", customers.filter(col("name").isNull()).count()),
    ("Negative Price", cars.filter(col("price") < 0).count()),
    ("Invalid Customers", invalid_customers.count())
], ["issue", "count"])

validation.show()
```

# Transformation

## Base Join

```python
from pyspark.sql.functions import col, sum

df = (
    sales_clean.alias("s")
    .join(customers.alias("c"), col("s.customer_id") == col("c.customer_id"))
    .join(cars.alias("car"), col("s.car_id") == col("car.car_id"))
)
```

## 1. Revenue per Customer

```python
revenue_df = (
    df.groupBy(col("c.customer_id"), col("c.name"), col("c.city"))
    .agg(sum(col("car.price") * col("s.quantity")).alias("total_revenue")).orderBy(col("total_revenue").desc()))
revenue_df.display()
```

## 2 Cars per brand

```python
brand_df = (cars.groupBy("brand").count().withColumnRenamed("count", "total_cars"))
brand_df.display()
```

## 3 City revenue

```python
city_df = (df.groupBy(col("c.city")).agg(sum(col("car.price") * col("s.quantity")).alias("city_revenue")))
city_df.display()
```

# Dealer Analytics

```python
dealer_df = (
    sales_clean.alias("s")
    .join(cars.alias("car"), col("s.car_id") == col("car.car_id"))
    .join(sales_dealer.alias("sd"), col("s.sale_id") == col("sd.sale_id"))
    .join(dealers.alias("d"), col("sd.dealer_id") == col("d.dealer_id"))
    .select(
        col("d.dealer_id"),
        col("d.name").alias("dealer_name"),
        col("d.city").alias("dealer_city"),
        col("car.price"),
        col("s.quantity")
    )
)
dealer_df.display()
```

## 1) Revenue per Dealer

```python
dealer_revenue_df = (
    dealer_df.groupBy("dealer_id", "dealer_name", "dealer_city")
    .agg(sum(col("price") * col("quantity")).alias("total_revenue"))
)
dealer_revenue_df.display()
```

## 2) Top Dealers by Revenue

```python
top_dealers_df = dealer_revenue_df.orderBy(desc("total_revenue"))
top_dealers_df.show(5)
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number

window_spec = Window.orderBy(col("total_revenue").desc())

top3_dealers = (
    dealer_revenue_df
    .withColumn("rn", row_number().over(window_spec))
    .filter(col("rn") <= 3)
)
top3_dealers.display()
```

## 3) Dealer-City Contribution

```python
city_total = dealer_revenue_df.groupBy("dealer_city") \
    .agg(sum("total_revenue").alias("city_total"))
dealer_contribution = (
    dealer_revenue_df.alias("d")
    .join(city_total.alias("c"), "dealer_city")
    .select(
        col("d.dealer_id"),
        col("d.dealer_name"),
        col("d.dealer_city"),
        col("d.total_revenue"),
        (col("d.total_revenue") / col("c.city_total") * 100).alias("contribution_pct")
    )
)
dealer_contribution.display()
```

# Analysis

```python
clean_df = df.select(
    col("s.sale_id"),
    col("s.customer_id"),
    col("c.name").alias("customer_name"),
    col("c.city").alias("customer_city"),
    col("car.brand"),
    col("car.price"),
    col("s.quantity"),
    col("s.sale_date")
)
```

## 1) Top 3 Customers per City

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import sum, row_number
customer_revenue = (
    clean_df.groupBy("customer_city", "customer_id", "customer_name")
    .agg(sum(col("price") * col("quantity")).alias("total_revenue"))
)
window_spec = Window.partitionBy("customer_city").orderBy(col("total_revenue").desc())
top3_df = (
    customer_revenue
    .withColumn("rn", row_number().over(window_spec))
    .filter(col("rn") <= 3)
)
top3_df.display()
```

## 2) Repeat Customers

```python
from pyspark.sql.functions import count
repeat_df = (
    clean_df.groupBy("customer_id")
    .agg(count("*").alias("total_orders"))
    .filter(col("total_orders") > 1)
)
repeat_df.display()
```

## 3) Monthly Trend

```python
from pyspark.sql.functions import date_format
monthly_df = (
    clean_df
    .withColumn("month", date_format(col("sale_date"), "yyyy-MM"))
    .groupBy("month")
    .agg(sum(col("price") * col("quantity")).alias("revenue"))
    .orderBy("month")
)
monthly_df.display()
```
